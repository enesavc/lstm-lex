```python
#ALL3Models_ACC Plot
from seaborn import lineplot

sns.set(font_scale = 3)
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize= (30,15))


red = "#D81B60"
navy = "#0C7BDC"
orange = "#FFC20A"

ax1 = lineplot(data=df_Dorsal, x=df_Dorsal.index, y=df_Dorsal.DorsalAVG, errorbar=None, color=navy,
              marker="o", markersize=10)
ax1.fill_between(df_Dorsal.index, df_Dorsal.DorsalLower, df_Dorsal.DorsalUpper, alpha=0.2, color=navy,
                label='Dorsal Train')

ax2 = lineplot(data=df_Dorsal_v, x=df_Dorsal_v.index, y=df_Dorsal_v.DorsalAVG, errorbar=None, color=navy,
              linestyle="dashed", marker="o", markersize=10)
ax2.fill_between(df_Dorsal_v.index, df_Dorsal_v.DorsalLower, df_Dorsal_v.DorsalUpper, alpha=0.2, color=navy,
                label='Dorsal Valid')

ax3 = lineplot(data=df_Fused, x=df_Fused.index, y=df_Fused.FusedAVG, errorbar=None, color=orange,
              marker="^", markersize=10)
ax3.fill_between(df_Fused.index, df_Fused.FusedLower, df_Fused.FusedUpper, alpha=0.2, color=orange,
                label='Fused Train')

ax4 = lineplot(data=df_Fused_v, x=df_Fused_v.index, y=df_Fused_v.FusedAVG, errorbar=None, color=orange,
              linestyle="dashed",  marker="^", markersize=10)
ax4.fill_between(df_Fused_v.index, df_Fused_v.FusedLower, df_Fused_v.FusedUpper, alpha=0.2, color=orange,
                label='Fused Valid')

ax5 = lineplot(data=df_Ventral, x=df_Ventral.index, y=df_Ventral.VentralAVG, errorbar=None, color=red,
              marker="s", markersize=10)
ax5.fill_between(df_Ventral.index, df_Ventral.VentralLower, df_Ventral.VentralUpper, alpha=0.2, color=red,
                label='Ventral Train')

ax6 = lineplot(data=df_Ventral_v, x=df_Ventral_v.index, y=df_Ventral_v.VentralAVG, errorbar=None, color=red,
              linestyle="dashed", marker="s", markersize=10)
ax6.fill_between(df_Ventral_v.index, df_Ventral_v.VentralLower, df_Ventral_v.VentralUpper, alpha=0.2, color=red,
                label='Ventral Valid')

ax.set(xlabel="Epoch", ylabel="Average Cosine Similarity")
ax.set_ylim([0.5, 1.01])
ax.set_xlim([0, 100])

import matplotlib.lines as mlines
import matplotlib.pyplot as plt

b1 = mlines.Line2D([], [], color=navy, marker='o',
                          markersize=15, label='Dorsal Train')
b2 = mlines.Line2D([], [], color=navy, marker='o',
                          markersize=15, linestyle="dashed", label='Dorsal Val.')
b3 = mlines.Line2D([], [], color=orange, marker='^',
                          markersize=15, label='Fused Train')
b4 = mlines.Line2D([], [], color=orange, marker='^',
                          markersize=15, linestyle="dashed", label='Fused Val.')
b5 = mlines.Line2D([], [], color=red, marker='s',
                          markersize=15, label='Ventral Train')
b6 = mlines.Line2D([], [], color=red, marker='s',
                          markersize=15, linestyle="dashed", label='Ventral Val.')
plt.legend(handles=[b1, b2, b3, b4, b5, b6], loc = 'lower center', bbox_to_anchor=(0.5, -0.2),
          fancybox=True, shadow=True, ncol=6)
plt.tight_layout()
#ax.legend(loc = 'lower right')
#ax.set_title('Model Train Accuracy')
plt.savefig('Fig2',dpi=300)
```

# Fig 3

```python
import chardet
import pandas as pd

with open('fig3_data.csv', 'rb') as f:
    enc = chardet.detect(f.read())  # or readline if the file is large
    
df = pd.read_csv('fig3_data.csv', encoding = enc['encoding'])
#df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
sns.set(font_scale = 3)
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize= (30,15))

colors = ["#0C7BDC", "#FFC20A", "#D81B60"]
  


# create plot
g = sns.barplot(x = 'Task', y = 'ACC', hue = 'Model', data = df,
            palette = colors,
            order = ['OPM', 'POS'],  
            capsize = 0.05,             
            saturation = 8,             
            errcolor = 'black', errwidth = 2,  
            errorbar = 'sd'   
            )
g.set(xticklabels=["Onset Phoneme Monitoring", "Part of Speech Categorization"])
ax.set(xlabel="Tasks", ylabel="Decoding Accuracy (AMI)")
ax.set_ylim([0, 0.5])

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -0.28),
          fancybox=True, shadow=True, ncol=3)
plt.tight_layout()

plt.savefig('Fig3',dpi=300)
```
