```python
#DF TRAIN
import pandas as pd
df_Fused= pd.DataFrame({'Fused1': fused1_t_avg_cos, 'Fused2': fused2_t_avg_cos,
                  'Fused3': fused3_t_avg_cos, 'Fused4': fused4_t_avg_cos,
                  'Fused5': fused5_t_avg_cos, 'Fused6': fused6_t_avg_cos, 'Fused7': fused7_t_avg_cos,
                  'Fused8': fused8_t_avg_cos, 'Fused9': fused9_t_avg_cos,
                  'Fused10': fused10_t_avg_cos},
                 columns=['Fused1', 'Fused2','Fused3', 'Fused4','Fused5',
                          'Fused6', 'Fused7','Fused8', 'Fused9','Fused10'])
df_Fused['FusedAVG'] = df_Fused.mean(axis=1)
df_Fused['FusedSTD'] = df_Fused.std(axis=1)
#df['fusedSEM'] = df.sem(axis=1)
df_Fused['FusedLower'] = df_Fused['FusedAVG'] - df_Fused['FusedSTD']
df_Fused['FusedUpper'] = df_Fused['FusedAVG'] + df_Fused['FusedSTD']
#df_Fused

#to store it
%store df_Fused
# del df_fused
```

```python
#DF VALID
import pandas as pd
df_Fused_v= pd.DataFrame({'Fused1': fused1_v_avg_cos, 'Fused2': fused2_v_avg_cos,
                  'Fused3': fused3_v_avg_cos, 'Fused4': fused4_v_avg_cos,
                  'Fused5': fused5_v_avg_cos, 'Fused6': fused6_v_avg_cos, 'Fused7': fused7_v_avg_cos,
                  'Fused8': fused8_v_avg_cos, 'Fused9': fused9_v_avg_cos,
                  'Fused10': fused10_v_avg_cos},
                 columns=['Fused1', 'Fused2','Fused3', 'Fused4','Fused5',
                          'Fused6', 'Fused7','Fused8', 'Fused9','Fused10'])
df_Fused_v['FusedAVG'] = df_Fused_v.mean(axis=1)
df_Fused_v['FusedSTD'] = df_Fused_v.std(axis=1)
#df['FusedSEM'] = df.sem(axis=1)
df_Fused_v['FusedLower'] = df_Fused_v['FusedAVG'] - df_Fused_v['FusedSTD']
df_Fused_v['FusedUpper'] = df_Fused_v['FusedAVG'] + df_Fused_v['FusedSTD']
#df_Fused_v

#to store it
%store df_Fused_v
# del df_Fused_v
```

```python
from seaborn import lineplot

sns.set_style("white")
#sns.set(font_scale = 2)
fig, ax = plt.subplots(figsize= (20,10))
ax = lineplot(data=df_Fused, x=df_Fused.index, y=df_Fused.FusedAVG, errorbar=None, color='orange')
ax.fill_between(df_Fused.index, df_Fused.FusedLower, df_Fused.FusedUpper, alpha=0.2, color='orange', label='Fused Train')

ax = lineplot(data=df_Fused_v, x=df_Fused_v.index, y=df_Fused_v.FusedAVG, errorbar=None, color='orange', linestyle="dashed")
ax.fill_between(df_Fused_v.index, df_Fused_v.FusedLower, df_Fused_v.FusedUpper, alpha=0.2, color='orange', label='Fused Valid')

ax.set(xlabel="Epoch", ylabel="Average Cosine Similarity")
ax.set_ylim([0, 1.05])
ax.set_title('Fused Model Train and Validation Accuracy')
ax.legend(loc = 'lower right')
#plt.savefig('AVG Training Accuracy by Epoch',dpi=300)
```
