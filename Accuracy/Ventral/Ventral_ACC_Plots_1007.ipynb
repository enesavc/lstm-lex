```python
#DF TRAIN
import pandas as pd
df_Ventral= pd.DataFrame({'Ventral1': ventral1_t_avg_cos, 'Ventral2': ventral2_t_avg_cos,
                  'Ventral3': ventral3_t_avg_cos, 'Ventral4': ventral4_t_avg_cos,
                  'Ventral5': ventral5_t_avg_cos, 'Ventral6': ventral6_t_avg_cos, 'Ventral7': ventral7_t_avg_cos,
                  'Ventral8': ventral8_t_avg_cos, 'Ventral9': ventral9_t_avg_cos,
                  'Ventral10': ventral10_t_avg_cos},
                 columns=['Ventral1', 'Ventral2','Ventral3', 'Ventral4','Ventral5',
                          'Ventral6', 'Ventral7','Ventral8', 'Ventral9','Ventral10'])
df_Ventral['VentralAVG'] = df_Ventral.mean(axis=1)
df_Ventral['VentralSTD'] = df_Ventral.std(axis=1)
#df['VentralSEM'] = df.sem(axis=1)
df_Ventral['VentralLower'] = df_Ventral['VentralAVG'] - df_Ventral['VentralSTD']
df_Ventral['VentralUpper'] = df_Ventral['VentralAVG'] + df_Ventral['VentralSTD']
#df_Ventral

#to store it
%store df_Ventral
# del df_Ventral
```

```python
#DF VALID
import pandas as pd
df_Ventral_v= pd.DataFrame({'Ventral1': ventral1_v_avg_cos, 'Ventral2': ventral2_v_avg_cos,
                  'Ventral3': ventral3_v_avg_cos, 'Ventral4': ventral4_v_avg_cos,
                  'Ventral5': ventral5_v_avg_cos, 'Ventral6': ventral6_v_avg_cos, 'Ventral7': ventral7_v_avg_cos,
                  'Ventral8': ventral8_v_avg_cos, 'Ventral9': ventral9_v_avg_cos,
                  'Ventral10': ventral10_v_avg_cos},
                 columns=['Ventral1', 'Ventral2','Ventral3', 'Ventral4','Ventral5',
                          'Ventral6', 'Ventral7','Ventral8', 'Ventral9','Ventral10'])
df_Ventral_v['VentralAVG'] = df_Ventral_v.mean(axis=1)
df_Ventral_v['VentralSTD'] = df_Ventral_v.std(axis=1)
#df['VentralSEM'] = df.sem(axis=1)
df_Ventral_v['VentralLower'] = df_Ventral_v['VentralAVG'] - df_Ventral_v['VentralSTD']
df_Ventral_v['VentralUpper'] = df_Ventral_v['VentralAVG'] + df_Ventral_v['VentralSTD']
#df_Ventral_v

#to store it
%store df_Ventral_v
# del df_Ventral_v
```

```python
from seaborn import lineplot

sns.set_style("white")
#sns.set(font_scale = 2)
fig, ax = plt.subplots(figsize= (20,10))
ax = lineplot(data=df_Ventral, x=df_Ventral.index, y=df_Ventral.VentralAVG, errorbar=None, color='red')
ax.fill_between(df_Ventral.index, df_Ventral.VentralLower, df_Ventral.VentralUpper, alpha=0.2, color='red', label='Ventral Train')

ax = lineplot(data=df_Ventral_v, x=df_Ventral_v.index, y=df_Ventral_v.VentralAVG, errorbar=None, color='red', linestyle="dashed")
ax.fill_between(df_Ventral_v.index, df_Ventral_v.VentralLower, df_Ventral_v.VentralUpper, alpha=0.2, color='red', label='Ventral Valid')

ax.set(xlabel="Epoch", ylabel="Average Cosine Similarity")
ax.set_ylim([0, 1.05])
ax.set_title('Ventral Model Train and Validation Accuracy')
ax.legend(loc = 'lower right')
#plt.savefig('AVG Training Accuracy by Epoch',dpi=300)
```
