```python
#DF TRAIN
import pandas as pd
df_Dorsal= pd.DataFrame({'Dorsal1': dorsal1_t_avg_cos, 'Dorsal2': dorsal2_t_avg_cos,
                  'Dorsal3': dorsal3_t_avg_cos, 'Dorsal4': dorsal4_t_avg_cos,
                  'Dorsal5': dorsal5_t_avg_cos, 'Dorsal6': dorsal6_t_avg_cos, 'Dorsal7': dorsal7_t_avg_cos,
                  'Dorsal8': dorsal8_t_avg_cos, 'Dorsal9': dorsal9_t_avg_cos,
                  'Dorsal10': dorsal10_t_avg_cos},
                 columns=['Dorsal1', 'Dorsal2','Dorsal3', 'Dorsal4','Dorsal5',
                          'Dorsal6', 'Dorsal7','Dorsal8', 'Dorsal9','Dorsal10'])
df_Dorsal['DorsalAVG'] = df_Dorsal.mean(axis=1)
df_Dorsal['DorsalSTD'] = df_Dorsal.std(axis=1)
#df['DorsalSEM'] = df.sem(axis=1)
df_Dorsal['DorsalLower'] = df_Dorsal['DorsalAVG'] - df_Dorsal['DorsalSTD']
df_Dorsal['DorsalUpper'] = df_Dorsal['DorsalAVG'] + df_Dorsal['DorsalSTD']
#df_Dorsal

#to store it
%store df_Dorsal
# del df_Dorsal
```

```python
#DF VALID
import pandas as pd
df_Dorsal_v= pd.DataFrame({'Dorsal1': dorsal1_v_avg_cos, 'Dorsal2': dorsal2_v_avg_cos,
                  'Dorsal3': dorsal3_v_avg_cos, 'Dorsal4': dorsal4_v_avg_cos,
                  'Dorsal5': dorsal5_v_avg_cos, 'Dorsal6': dorsal6_v_avg_cos, 'Dorsal7': dorsal7_v_avg_cos,
                  'Dorsal8': dorsal8_v_avg_cos, 'Dorsal9': dorsal9_v_avg_cos,
                  'Dorsal10': dorsal10_v_avg_cos},
                 columns=['Dorsal1', 'Dorsal2','Dorsal3', 'Dorsal4','Dorsal5',
                          'Dorsal6', 'Dorsal7','Dorsal8', 'Dorsal9','Dorsal10'])
df_Dorsal_v['DorsalAVG'] = df_Dorsal_v.mean(axis=1)
df_Dorsal_v['DorsalSTD'] = df_Dorsal_v.std(axis=1)
#df['DorsalSEM'] = df.sem(axis=1)
df_Dorsal_v['DorsalLower'] = df_Dorsal_v['DorsalAVG'] - df_Dorsal_v['DorsalSTD']
df_Dorsal_v['DorsalUpper'] = df_Dorsal_v['DorsalAVG'] + df_Dorsal_v['DorsalSTD']
#df_Dorsal_v

#to store it
%store df_Dorsal_v
# del df_Dorsal_v
```

```python
from seaborn import lineplot

sns.set_style("white")
#sns.set(font_scale = 2)
fig, ax = plt.subplots(figsize= (20,10))
ax = lineplot(data=df_Dorsal, x=df_Dorsal.index, y=df_Dorsal.DorsalAVG, errorbar=None, color='navy')
ax.fill_between(df_Dorsal.index, df_Dorsal.DorsalLower, df_Dorsal.DorsalUpper, alpha=0.2, color='navy', label='Dorsal Train')

ax = lineplot(data=df_Dorsal_v, x=df_Dorsal_v.index, y=df_Dorsal_v.DorsalAVG, errorbar=None, color='navy', linestyle="dashed")
ax.fill_between(df_Dorsal_v.index, df_Dorsal_v.DorsalLower, df_Dorsal_v.DorsalUpper, alpha=0.2, color='navy', label='Dorsal Valid')

ax.set(xlabel="Epoch", ylabel="Average Cosine Similarity")
ax.set_ylim([0, 1.05])
ax.set_title('Dorsal Model Train and Validation Accuracy')
ax.legend(loc = 'lower right')
#plt.savefig('AVG Training Accuracy by Epoch',dpi=300)
```
